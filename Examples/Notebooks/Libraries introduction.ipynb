## Loading data

Data will be loaded first using numpy, as the handling is far simpler. Later, data will be loaded using Pandas for a more complete analysis of the data series.

```python
#using numpy
f_data = 'brue_data.csv'

# have to keep in mind [Rt, Et, Ret, Qt]
data_np = np.loadtxt(f_data, delimiter=',', skiprows=1, usecols=[1, 2, 3, 4])

# parsing precipitation data
prec = data_np[:, 0]

# Plotting precipitation
plt.figure()  # Optional to create a new figure
plt.plot(prec, label='Rainfall [mm]', marker=None, color='b')  # Make the plot
plt.legend()  # Display legend
plt.grid()  # Display grid on ticks
plt.show()  # Display figure
```

```python
prec[200] = -9999
```

```python
## Use this space to make the plot of the other variables
# Consider using different colors, subplots and different markers

# Plotting prec event
plt.figure()
plt.subplot(2,1,1)
plt.plot(prec)
plt.xlim(3000, 3500)
plt.ylim(0, 8)

plt.subplot(2,1,2)
plt.plot(data_np[:,-1])
plt.xlim(3000, 3500)
plt.show()


p80 = np.percentile(prec, 80)
print(p80)

plt.hist(prec, log=True, bins=50)
plt.axvline(p80, color='r')
plt.show()
```

```python
prec[prec < 0] = 0
```

```python
## Making some statistics
#

#getting some simple indicators
average_prec = np.average(prec)
std_prec = np.std(prec)
cov_prec = std_prec / average_prec

# Printing with some fancy formatting
print('Precipitation mean: {0:.2f} and precipitation standard deviation: {1:.2f}'.
      format(average_prec, std_prec))
print('Precipitation cov: {0:.3f}'.format(cov_prec))

# Iterating over the array to make autocorrelation plot
n_lags = 20  # Number of steps for autocorrelation function

prec_autocorr = [1.0, ]
for lag in xrange(1, n_lags + 1):  # xrange changed for range in python 3
    temp_acorr = np.corrcoef(prec[:-lag], prec[lag:])[1,0] # autocorrelation for lags larger than 0
    prec_autocorr.append(temp_acorr)

# Plotting the results of the autocorrelation analysis
plt.figure()
plt.plot(prec_autocorr, label='Prec', marker='d')
plt.ylabel('Autocorrelation')
plt.xlabel('lag [hr]')
plt.legend()
plt.grid()
plt.show()
```

## Using pandas to manage the data and doing some fancy statistics

Data is being loaded into a pandas DataFrame (df). This df is being used to manipulate the information in different ways. One of the main features of panda is the ability to efficiently work with large ammounts of data. Also, is possible to parse dates, and do temporal agregation of the datasets.


```python
# Load dataset using pandas

# Defining from the CSv files the initial and end dates
data_start = '24-06-1994 05:00'
data_end = '24-06-1995 04:00'

# Make the vector with the index of time stamps (from start to end every hour)
time_index = pd.date_range(data_start, data_end, freq='H')

# Load the data from the file skipping the first column (strange date)
data_pd = pd.read_csv(f_data, usecols=[1, 2, 3, 4])

# Modify (in place) the dataset to set the index with the time_index
data_pd.set_index(time_index, inplace=True)

# Display the first few data
# data_pd.head()
data_pd.tail()
```

```python
# Make a plot of the precipitation data using pandas
data_pd['Rt'].plot()  # Index to access the data is the Key (Rt)
plt.grid()
plt.legend()
plt.ylabel('Precipitation [mm]')
plt.show()
```

```python
# Here we resample the data in 2 ways: total and average precipitation

# Make plots resampling at different temporal scales (total precipitaiton in interval)
plt.figure()
data_pd['Rt'].resample('D').sum().plot()
data_pd['Rt'].resample('M').sum().plot()
data_pd['Rt'].resample('3M').sum().plot()
plt.grid()
plt.legend()
plt.ylabel('Precipitation [mm]')
plt.title('Total precipitation')
plt.show()

# Make plots resampling at different temporal scales (average precipitaiton in interval)
plt.figure()
data_pd['Rt'].resample('D').mean().plot()
data_pd['Rt'].resample('M').mean().plot()
data_pd['Rt'].resample('3M').mean().plot()
plt.grid()
plt.legend()
plt.ylabel('Precipitation [mm]')
plt.title('Average precipitation')
plt.show()
```
